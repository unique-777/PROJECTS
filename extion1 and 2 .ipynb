```python
x = data.drop("age", axis = 1)
y = data["age"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.7)
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
import autokeras as ak
clf = ak.StructuredDataClassifier(
    max_trials=10,overwrite=True
)
clf.fit(x_train, y_train, epochs=10)
accuracy = clf.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy}")
```

```python
best= clf.export_model()
print("Best Model Summary:")
best.summary()
y_pred = clf.predict(x_test)
```

```python
best.save("autokeras.model")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
label_encoder = LabelEncoder()
y_test_encoded = label_encoder.fit_transform(y_test)
y_pred_encoded = label_encoder.transform(y_pred)
print("Classification Report:")
print(classification_report(y_test_encoded, y_pred_encoded))
print("Confusion Matrix:")
print(confusion_matrix(y_test_encoded, y_pred_encoded))
```

```python
y_test_num = y_test.astype(int)
y_pred_num = y_pred.astype(int)
print("Classification Report:")
print(classification_report(y_test_num, y_pred_num))
print("Confusion Matrix:")
print(confusion_matrix(y_test_num, y_pred_num))
```

```python
from tensorflow.keras import layers, models
def create_prototypical_network(input_shape):
    input = layers.Input(shape=input_shape)
    x = layers.Flatten()(input)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(32, activation='relu')(x)    
    output = layers.Dense(10)(x)  
    model = models.Model(inputs=input, outputs=output)
    return model
input_shape = (28, 28)  
model = create_prototypical_network(input_shape)
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum(np.square(a - b), axis=-1))
```

```python
def train_prototypical_network(model, x_train, y_train, n_epochs=10, n_way=5, n_shot=5):
    for epoch in range(n_epochs):
        print(f'Epoch {epoch + 1}/{n_epochs}')
        for class_ in np.unique(y_train):
            class_data = x_train[y_train == class_]
            if len(class_data) < n_shot:
                continue
            support_set = class_data[np.random.choice(class_data.shape[0], size=n_shot, replace=False)]
            support_set_embeddings = model.predict(support_set)
            prototype = np.mean(support_set_embeddings, axis=0)
            query_set = class_data  
            query_set_embeddings = model.predict(query_set)
            distances = euclidean_distance(query_set_embeddings, prototype)
            predicted_class = np.argmin(distances, axis=0)

        print(f'Epoch {epoch + 1} complete.')
input_shape = (28, 28, 1)
model = create_prototypical_network(input_shape)
train_prototypical_network(model, x_train, y_train, n_epochs=10)
```
